## concat

* 가로 혹은 세로로 붙일 수 있음
* pandas.concat(objs, axis=0, join='outer', join_axes=None, ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=None, copy=True)
* DataFrame에 컬럼 및 행을 추가할 때 사용함
* pd.concat()의 매개변수로 axis 값을 1로 지정하면 열, 0으로 지정하면 행으로 처리
* concat는 concatenation의 단어에서 따옴

```python
sk = "https://finance.naver.com/item/sise_day.nhn?code=096770&page={}"
original_sk_df = pd.DataFrame()
for x in range(1, 6):
    original_sk_df = pd.concat([original_sk_df, pd.read_html(sk.format(x) )[0]])
```

```python
original_sk_df
```

```python
original_sk_df.dropna(inplace=True)
```

## merge

* copy() : df 할당은 데이터 할당이 아님. 데이터를 할당하려면 copy()를 사용해야 함

```python
wti = original.copy()
```

```python
tmp = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
              'B': ['B0', 'B1', 'B2', 'B3'],
              'C': ['C0', 'C1', 'C2', 'C3'],
              'D': ['D0', 'D1', 'D2', 'D3']},
              index=[0, 1, 2, 3])
```

```python
tmp
```

```python
tmp2 = tmp.iloc[2:, :]
tmp2
```

```python
tmp2.iloc[1,3]
```

```python
tmp2.iloc[1,3] = 'X'
```

```python
tmp2
```

```python
tmp
```

```python
s1 = tmp.iloc[2:, :].copy()
```

```python
s1.iloc[1,3] = 'Y'
```

```python
s1
```

```python
tmp
```

* copy()는 원본에 영향을 미치지 않는 독립된 새로운 df를 만드는 방법
* 그렇지 않고, 기존 df에서 일부를 할당할 경우, 할당된 데이터의 변경이 기존 df에 영향을 미치게 됨

```python
original_wti.head(3)
```

```python
original_sk_df.head(3)
```

```python
original_wti.merge(original_sk_df)
```

```python
original_wti.merge(original_sk_df, on = ['날짜'])
```

```python
original_wti.merge(original_sk_df, how = 'outer', on = ['날짜'])
```
